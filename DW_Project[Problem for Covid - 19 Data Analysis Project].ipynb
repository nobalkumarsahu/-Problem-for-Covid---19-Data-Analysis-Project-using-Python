# 4. Filter the dataframe with only this columns 
     ['continent','location','date','total_cases','total_deaths','gdp_per_capita',' human_development_index'] 
     and update the data frame.

```python
new_data= data.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
new_data
```

# 5. Data Cleaning 
        a. Remove all duplicates observations 
        b. Find missing values in all columns 
        c. Remove all observations where continent column value is missing Tip : using subset parameter in dropna 
        d. Fill all missing values with 0

```python
new_data.drop_duplicates(inplace=True)
new_data
```

```python
new_data.isna()
```

```python
new_data=new_data.fillna(0)
new_data
```

```python
new_data=new_data.loc[-((new_data['continent'] == 0))]
new_data
```

# 6. Date time format : 
        a. Convert date column in datetime format using pandas.to_datetime
        b. Create new column month after extracting month data from date column.

```python
new_data['date']=pd.to_datetime(new_data['date'])
new_data.info()
```

```python
new_data['month'] = pd.DatetimeIndex(new_data['date']).month
new_data
```

# 7. Data Aggregation:
    a. Find max value in all columns using groupby function on 'continent' column 
        Tip: use reset_index() after applying groupby
    b. Store the result in a new dataframe named 'df_groupby'.
        (Use df_groupby dataframe for all further analysis)

```python
df_groupby=new_data.groupby('continent')[['location','date','total_cases','total_deaths','gdp_per_capita','human_development_index','month']].max().reset_index()
```

```python
df_groupby
```

# 8. Feature Engineering :
     a. Create a new feature 'total_deaths_to_total_cases' by ratio of
     'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
df_groupby
```
